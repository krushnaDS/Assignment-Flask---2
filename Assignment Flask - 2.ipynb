```python
'''
Q1. Explain GET and POST methods.

GET is used to retrieve data from the server. It is the most common method for fetching resources, 
such as web pages, images, and files. The data that is retrieved is usually displayed in the browser.

POST is used to submit data to the server. It is often used for forms, where the user enters data 
that is then sent to the server for processing. The data that is submitted is not usually displayed in the browser.

from flask import Flask, request

app = Flask(__name__)

@app.route("/get", methods=["GET"])
def get():
    name = request.args.get("name")
    return f"Hello, {name}!"

@app.route("/post", methods=["POST"])
def post():
    name = request.form["name"]
    return f"Hello, {name}! Your message has been submitted."

if __name__ == "__main__":
    app.run()
'''
```

```python
'''
Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect the user to a different URL. This can be useful for a variety of reasons, such as:

To prevent the user from seeing a broken or outdated page.
To send the user to a different page after they have logged in or out.
To redirect the user to a different page after they have submitted a form.
To implement 301 or 302 redirects.
The redirect() function takes two arguments: the URL to redirect the user to, and the status code of the redirect. 
The default status code is 302, which means that the redirect is temporary. However, you can also specify other 
status codes, such as 301 (permanent redirect) or 308 (multiple choice redirect).

'''
```
